```python
import codecs
```

```python
singleTEI_r2 = codecs.open(os.path.join("/Users/lulinghuang/Desktop/2017 DSC/eb_project_debug/debug_090117_eb09_r02/2-tei", "170711-eb09-r02-0001.xml"), 'r', encoding='utf-8').read()
```

```python
singleTEI_soup_r2 = BeautifulSoup(singleTEI_r2, "lxml-xml")
```

```python
singleTEI_r1 = codecs.open(os.path.join("/Users/lulinghuang/Desktop/2017 DSC/eb_project_debug/debug_090117_eb09_r01/2-tei", "170705-eb09-r01-0001.xml"), 'r', encoding='utf-8').read()
```

```python
singleTEI_soup_r1 = BeautifulSoup(singleTEI_r1, "lxml-xml")
```

```python
invalid_tags = ['body', 'fileDesc', 'teiHeader', 'text', 'lb', 'TEI', 'titleStmt', 'title', 'author', 'publicationStmt', 'sourceDesc','p/']
for tag in invalid_tags: 
    for match in singleTEI_soup_r1.findAll(tag):
        match.replaceWithChildren()
page_r1 = singleTEI_soup_r1.prettify(formatter = None)
```

```python
print (page_r1)
```

```python
invalid_tags = ['body', 'fileDesc', 'teiHeader', 'text', 'lb', 'TEI', 'titleStmt', 'title', 'author', 'publicationStmt', 'sourceDesc','p/']
for tag in invalid_tags: 
    for match in singleTEI_soup_r2.find_all(tag):
        match.unwrap()
page_r2 = singleTEI_soup_r2.prettify(formatter = None)
```

```python
print (page_r2)
```

```python
singleTEI_soup_r2.find_all(rend = 'font2')
```

```python
invalid_hi_tags = ['font2', 'font1', 'italic']
for tag in invalid_hi_tags: 
    for match in singleTEI_soup_r2.find_all(rend = tag):
        match.unwrap()
page_r2 = singleTEI_soup_r2.prettify(formatter = None)
```

```python
print (page_r2)
```
